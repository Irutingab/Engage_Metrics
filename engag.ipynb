# Academic Performance Analysis

Let's create a detailed bar chart showing how many students achieve different grade thresholds (50+, 60+, 70+, 80+, 90+).

```python
# Import the Visualizations class from our specialized file
from visualizations import Visualizations
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('StudentPerformanceFactors.csv')

# Use the Visualizations class method (following OOP principles)
fig = Visualizations.create_academic_performance_threshold_chart(df)
plt.show()

# Print detailed statistics
print("Academic Performance Statistics:")
print("=" * 40)
for threshold in [50, 60, 70, 80, 90]:
    count = (df['Exam_Score'] >= threshold).sum()
    percentage = (count / len(df)) * 100
    print(f"Students scoring {threshold}+: {count} ({percentage:.1f}%)")

print(f"\nTotal students: {len(df)}")
print(f"Average score: {df['Exam_Score'].mean():.1f}")
print(f"Median score: {df['Exam_Score'].median():.1f}")
print(f"Standard deviation: {df['Exam_Score'].std():.1f}")
```

```python
# Let's investigate the data quality issue - students with grades > 100
import pandas as pd

# Load the dataset
df = pd.read_csv('StudentPerformanceFactors.csv')

# Find students with exam scores greater than 100
students_over_100 = df[df['Exam_Score'] > 100]

print("Students with exam scores greater than 100:")
print("=" * 50)
print(f"Number of students: {len(students_over_100)}")
print()

if len(students_over_100) > 0:
    print("Details of students with scores > 100:")
    for idx, row in students_over_100.iterrows():
        print(f"Row {idx + 1}: Exam Score = {row['Exam_Score']}")
        print(f"  Other details: {row.to_dict()}")
        print()
    
    # Show basic statistics for comparison
    print("\nDataset Statistics for Exam_Score:")
    print(f"Minimum score: {df['Exam_Score'].min()}")
    print(f"Maximum score: {df['Exam_Score'].max()}")
    print(f"Mean score: {df['Exam_Score'].mean():.2f}")
    print(f"Standard deviation: {df['Exam_Score'].std():.2f}")
    
    # Show distribution of scores
    print(f"\nScore distribution:")
    score_ranges = [(0, 50), (51, 60), (61, 70), (71, 80), (81, 90), (91, 100), (101, float('inf'))]
    for min_score, max_score in score_ranges:
        if max_score == float('inf'):
            count = len(df[df['Exam_Score'] > min_score])
            print(f"Scores > {min_score}: {count} students")
        else:
            count = len(df[(df['Exam_Score'] >= min_score) & (df['Exam_Score'] <= max_score)])
            print(f"Scores {min_score}-{max_score}: {count} students")
else:
    print("No students found with exam scores greater than 100.")
```

```python
# Options for handling the outlier (score = 101)
print("Options for handling the student with score = 101:")
print("=" * 50)
print("1. Cap the score at 100 (most common approach)")
print("2. Remove the student record entirely")
print("3. Keep as is (if scores can legitimately exceed 100)")
print("4. Check if it's a data entry error and correct it")
print()

# Let's examine this student's profile more closely
outlier_student = df[df['Exam_Score'] > 100].iloc[0]
print("Detailed profile of the outlier student:")
print("=" * 40)
for column, value in outlier_student.items():
    print(f"{column}: {value}")

print(f"\nThis student has some interesting characteristics:")
print(f"- Studied for {outlier_student['Hours_Studied']} hours")
print(f"- Has {outlier_student['Attendance']}% attendance")
print(f"- Previous scores: {outlier_student['Previous_Scores']}")
print(f"- But parental involvement is {outlier_student['Parental_Involvement']}")
print(f"- And motivation level is {outlier_student['Motivation_Level']}")

# Option 1: Cap at 100 (demonstration)
print(f"\nIf we cap the score at 100:")
df_capped = df.copy()
df_capped.loc[df_capped['Exam_Score'] > 100, 'Exam_Score'] = 100
print(f"New maximum score: {df_capped['Exam_Score'].max()}")
print(f"Number of students with score = 100: {(df_capped['Exam_Score'] == 100).sum()}")
```

```python
# Remove the student with exam score > 100
import pandas as pd

# Load the original dataset
df = pd.read_csv('StudentPerformanceFactors.csv')

print("Before removing outlier:")
print(f"Total students: {len(df)}")
print(f"Maximum score: {df['Exam_Score'].max()}")
print(f"Students with score > 100: {len(df[df['Exam_Score'] > 100])}")

# Remove students with exam scores greater than 100
df_cleaned = df[df['Exam_Score'] <= 100].copy()

print(f"\nAfter removing outlier:")
print(f"Total students: {len(df_cleaned)}")
print(f"Maximum score: {df_cleaned['Exam_Score'].max()}")
print(f"Students removed: {len(df) - len(df_cleaned)}")

# Save the cleaned dataset
df_cleaned.to_csv('StudentPerformanceFactors_cleaned.csv', index=False)

print(f"\nCleaned dataset saved as 'StudentPerformanceFactors_cleaned.csv'")
print(f"Dataset now contains {len(df_cleaned)} students with scores between {df_cleaned['Exam_Score'].min()} and {df_cleaned['Exam_Score'].max()}")

# Show updated statistics
print(f"\nUpdated Statistics:")
print(f"Mean score: {df_cleaned['Exam_Score'].mean():.2f}")
print(f"Median score: {df_cleaned['Exam_Score'].median():.2f}")
print(f"Standard deviation: {df_cleaned['Exam_Score'].std():.2f}")
```

```python
# Verification: Check that the outlier has been completely removed
df_verification = pd.read_csv('StudentPerformanceFactors_cleaned.csv')

print("Verification of cleaned dataset:")
print("=" * 40)
print(f"Total students: {len(df_verification)}")
print(f"Minimum score: {df_verification['Exam_Score'].min()}")
print(f"Maximum score: {df_verification['Exam_Score'].max()}")

# Double-check for any scores > 100
outliers_remaining = df_verification[df_verification['Exam_Score'] > 100]
print(f"Students with scores > 100: {len(outliers_remaining)}")

if len(outliers_remaining) == 0:
    print("✅ SUCCESS: No students with scores > 100 found in cleaned dataset!")
else:
    print("❌ WARNING: Still found students with scores > 100!")

# Show score distribution for the cleaned dataset
print(f"\nScore distribution in cleaned dataset:")
score_ranges = [(50, 60), (61, 70), (71, 80), (81, 90), (91, 100)]
for min_score, max_score in score_ranges:
    count = len(df_verification[(df_verification['Exam_Score'] >= min_score) & (df_verification['Exam_Score'] <= max_score)])
    percentage = (count / len(df_verification)) * 100
    print(f"Scores {min_score}-{max_score}: {count} students ({percentage:.1f}%)")
```
